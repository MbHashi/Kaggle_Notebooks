```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

titanic_file_path_train = '/kaggle/input/titanic/train.csv'


titanic_train = pd.read_csv(titanic_file_path_train)
#Filter to find any missing values.
filtered_titanic_train = titanic_train.dropna(axis=1)
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()
data_with_imputed_values = my_imputer.fit_transform(titanic_train)

#describe to read the data to organize the features.
titanic_train.describe()

#Features are the headers.
titanic_features = ['PassengerId', 'Survived', 'Pclass', 'Age', 
                        'SibSp', 'Parch', 'Fare']
#Set the X & Y values.
y = titanic_train.Survived
X = titanic_train[titanic_features]

#split into train & val.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)


#Make model
titanic_model = RandomForestRegressor(random_state=1)
titanic_model.fit(train_X, train_y)

#Calculate MAE
preds_val = titanic_model.predict(val_X)
mae = mean_absolute_error(val_y, preds_val)
return(mae)
```
